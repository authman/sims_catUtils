This notebook will demonstrate how to read your own data into CatSim in a way that the CatSim classes can understand.

<b>CatSim database philosphy</b>

Catalogs in CatSim are generated by the `InstanceCatalog` class defined in the `sims_catalogs_measures` package.  Instance Catalogs get their data from instantiations of the `CatalogDBObject` class defined in the `sims_catalogs_generation` package.  `CatalogDBObject` connects to a database, constructs SQL queries on that database, and formats the query results in a way that `InstanceCatalog` expects.  Generally, if you want to read your own data into CatSim, you will need to

* (1) Format your data into a database file.


* (2) Write a daughter class of `CatalogDBObject` that knows how to interpret the your database file.

If your data exists in a raw text file (i.e. a catalog), the class `fileDBObject` defined in the `sims_catalogs_generation` package is capable of reading in a text file and converting it into a database.

We will first illustrate how to use `fileDBObject` to convert a text file into a database file.  We will then illustrated how to write a `CatalogDBObject` daughter class to interpret an arbitrary database file.

<b>Using `fileDBObject` to read in text file catalogs </b>

Suppose you have a catalog in a text file, like the one below

```python
!cat 'tutorialData/fakeCatalog.txt'
```

```python
import numpy
import os
from lsst.sims.catalogs.generation.db import fileDBObject

class fakeFileDBObject(fileDBObject):
    idColKey = 'id'
```

```python
dbName = 'myFakeDatabase.db'

if os.path.exists(dbName):
    os.unlink(dbName)

db = fakeFileDBObject('tutorialData/fakeCatalog.txt', runtable='testStars',
                      database=dbName)
```

```python
print db.show_mapped_columns()
```

```python
query = 'select * from testStars'
results = db.execute_arbitrary(query)
for line in results:
    print line
```

```python
print results['sedFile']
```

<b>Writing a `CatalogDBObject` daughter class</b>

```python
from lsst.sims.catalogs.generation.db import CatalogDBObject

class fakeCatalogDBObject(CatalogDBObject):
    idColKey = 'id'
    tableid = 'testStars'
    
    columns = [('raJ2000', 'rightAscension*PI()/180.0', numpy.float),
               ('decJ2000', 'declination*PI()/180.0', numpy.float),
               ('magNorm', 'magnitudeNormalization', numpy.float),
               ('sedFilename', 'sedFile', str, 256)]
```

```python
otherDb = fakeCatalogDBObject(database=dbName, driver='sqlite')
```

```python
otherDb.show_mapped_columns()
```

```python
colnames = ['id', 'raJ2000', 'sedFilename']
results = otherDb.query_columns(colnames=colnames)
for line in results:
    print line
```
