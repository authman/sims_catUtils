This notebook will demonstrate how to read your own data into CatSim in a way that the CatSim classes can understand.

<b>CatSim database philosphy</b>

Catalogs in CatSim are generated by the `InstanceCatalog` class defined in the `sims_catalogs_measures` package.  Instance Catalogs get their data from instantiations of the `CatalogDBObject` class defined in the `sims_catalogs_generation` package.  `CatalogDBObject` connects to a database, constructs SQL queries on that database, and formats the query results in a way that `InstanceCatalog` expects.  Generally, if you want to read your own data into CatSim, you will need to

* (1) Format your data into a database file.


* (2) Write a daughter class of `CatalogDBObject` that knows how to interpret the your database file.

If your data exists in a raw text file (i.e. a catalog), the class `fileDBObject` defined in the `sims_catalogs_generation` package is capable of reading in a text file and converting it into a database.

We will first illustrate how to use `fileDBObject` to convert a text file into a database file.  We will then illustrated how to write a `CatalogDBObject` daughter class to interpret an arbitrary database file.

<b>Using `fileDBObject` to read in text file catalogs </b>

Suppose you have a catalog in a text file, like the one below

```python
!cat 'tutorialData/fakeCatalog.txt'
```

In order to read this catalog in, you will use the class `fileDBObject`. `fileDBObject` reads in the catalog using `numpy.genfromtxt()` and then stores the data in a database file using `sqlalchemy`.  The minimal arguments you need to pass to the `fileDBObject` constructor are

* The path to the catalog file


* Thename of the database you want to create (this is the kwarg `database`)


* The name of the table in that database that you want to contain your catalog (this is the kwarg `runtable`)


* The name of the column that uniquely identifies each object in the catalog (this is the kwarg `idColKey`; note that your catalog needs to have a column which uniquely identifies each object)

This will result in a `sqlite` database of the name specified.  To create a database run by a different driver, specify the kwargs `driver`, `host`, and `port`.

```python
import numpy
import os
from lsst.sims.catalogs.generation.db import fileDBObject

help(fileDBObject.__init__)
```

```python
dbName = 'myFakeDatabase.db'

if os.path.exists(dbName):
    os.unlink(dbName)

db = fileDBObject('tutorialData/fakeCatalog.txt', runtable='testStars',
                      database=dbName, idColKey='id')
```

We now have a database `myFakeDatabase.db` containing the columns specified in our fake catalog.

```python
!ls myFake*.db
```

```python
print db.show_mapped_columns()
```

Because `fileDBObject` is a daughter class of `CatalogDBObject`, you can call any of the `CatalogDBObject` or `DBObject` member functsion on `db`.  For example, you can execute an arbitrary SQL query.

```python
query = 'select * from testStars'
results = db.execute_arbitrary(query)
for line in results:
    print line
```

```python
print results['sedFile']
```

This is helpful, but it is not very useful if you want to use the CatSim infrastructure, since CatSim Instance Catalogs expect very specific file names and formats, to which this database does not conform.  To transform the contents of this database into something that Instance Catalogs can use, one must write a `CatalogDBObject` daughter class that connects to `myFakeDatabase.db` and applied transformations to its columns.  We demonstrate this below.

(You may be wondering why, since `fileDBObject` is a daughter of `CatalogDBObject`, can we not just read in the database and apply the transformations with a `fileDBObject` daughter class.  We can.  We are splitting up the steps to make everything a bit more transparent.  In the final section of this notebook, we will roll both steps into one using a `fileDBObject` daughter class.)

<b>Writing a `CatalogDBObject` daughter class</b>

Now that we have read our catalog into a database (or maybe our data was in a database to begin with) we want to apply formatting to that database which CatSim's Instance Catalogs expect.  For example, our fake catalog contains the columns `rightAscension` (in degrees), `declination` (in degrees), `magnitudeNormalization`, and `sedFile`.  CatSim's Instance Catalog classes expect columns `raJ2000` (in radians), `decJ2000` (in radians), `magNorm`, and `sedFilename`.  We can apply the transformations from one set of columns to the other by defining a `CatalogDBObject` daughter class.

```python
from lsst.sims.catalogs.generation.db import CatalogDBObject

class fakeCatalogDBObject(CatalogDBObject):
    idColKey = 'id'
    tableid = 'testStars'
    
    columns = [('raJ2000', 'rightAscension*PI()/180.0', numpy.float),
               ('decJ2000', 'declination*PI()/180.0', numpy.float),
               ('magNorm', 'magnitudeNormalization', numpy.float),
               ('sedFilename', 'sedFile', str, 256)]
```

<b>Note:</b> `CatalogDBObject` has a meta class that keeps a registry of all daughter `CatalogDBObject` classes that have been defined.  This means that running the above cell more that once (i.e. re-defining a class you have already defined) will raise an error.  If you want to run the above cell more than once, re-start your kernel before hand.

When defining a daughter class of `CatalogDBObject`, we have to specify both the `idColKey` (as in `fileDBObject`) and the name of the database table we want to connect to (we will specify the actual database connection when we call the constructor).

We also have the option of defining the member variable `columns`.  `columns` is a list of tuples which specifies how to transform the columns native to the database into the columns expected by CatSim.  The form of the tuple is:

    (nameExpectedByCatSim, nameNativeToDB, dataType)

With this information, we can now connect to `myFakeDatabase.db` using our class `fakeCatalogDBObject`.

```python
mappedDb = fakeCatalogDBObject(database=dbName, driver='sqlite')
```

Calling `show_mapped_columns()`, we see that `mappedDb` contains knowledge of both the columns natively defined in the database, as well as their mappings.

```python
mappedDb.show_mapped_columns()
```

We can query the mapped columns using the `CatalogDBObject` method `query_columns()`.  <b>Note:</b> `execute_arbitrary()`, which we used to query the database in the `fileDBObject` example, has no knowledge of the mapped columns, since it executes a SQL query directly on `myFakeDatabase.db`.

```python
colnames = ['id', 'raJ2000', 'sedFilename', 'sedFile']
results = mappedDb.query_columns(colnames=colnames)
for line in results:
    print line
```

`mappedDb` is of a form that we can feed to an `InstanceCatalog` daughter class.

```python
from lsst.sims.catalogs.measures.instance import InstanceCatalog

class exampleCatalogClass(InstanceCatalog):
    column_outputs = ['id', 'raJ2000', 'decJ2000', 'magNorm', 'sedFilename']
```

<b>Note:</b> For the same reason as with the `CatalogDBObject` definition above, calling this cell more than once will raise an error.

```python
catalog = exampleCatalogClass(mappedDb)
catalog.write_catalog('mappedCatalog.txt')
```

```python
!cat 'tutorialData/fakeCatalog.txt'
!cat 'mappedCatalog.txt'
```

<b>Recall</b> that, with our mapped catalog, we transformed RA and Dec from degrees to radians.

<b>Doing everything with `fileDBObject`</b>

As promised, because `fileDBObject` is itself a daughter of `CatalogDBObject`, we can achieve the entire process by defining a `fileDBObject` daughter class that performs the column transformations, without the need of a separate `CatalogDBObject` class.

```python
class wonderCatalogDBObject(fileDBObject):
    idColKey = 'id'
    tableid = 'testStars'
    
    columns = [('raJ2000', 'rightAscension*PI()/180.0', numpy.float),
               ('decJ2000', 'declination*PI()/180.0', numpy.float),
               ('magNorm', 'magnitudeNormalization', numpy.float),
               ('sedFilename', 'sedFile', str, 256)]
```

```python
dbName = 'myFakeDatabase.db'

if os.path.exists(dbName):
    os.unlink(dbName)

wonderDb = wonderCatalogDBObject('tutorialData/fakeCatalog.txt', runtable='testStars',
                      database=dbName, idColKey='id')
```

```python
query = 'select * from testStars'
results = wonderDb.execute_arbitrary(query)
for line in results:
    print line
```

```python
colnames = ['id', 'raJ2000', 'sedFilename', 'sedFile']
results = wonderDb.query_columns(colnames=colnames)
for line in results:
    print line
```

```python
catalog = exampleCatalogClass(wonderDb)
catalog.write_catalog('wonderCatalog.txt')
```

```python
!cat 'tutorialData/fakeCatalog.txt'
!cat 'wonderCatalog.txt'
```
